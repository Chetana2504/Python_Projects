```python
# Q 19)  Is rainfall correlated with temperature? (Scatter + correlation)
plt.scatter(df['Rainfall'], df['Temperature'])
plt.xlabel("Temperture")
plt.ylabel("Rainfall")
plt.title("Scatter plot of Rainfall vs Temperture")
plt.show()
```

```python
df['Rainfall'].corr(df['Temperature'])
```

To check the relationship between rainfall and temperature, I plotted a scatter plot and calculated the correlation coefficient. The scatter plot shows the spread of points, and the correlation value tells the strength and direction of the relationship. A value close to 0 means weak or no correlation, while values near +1 or â€“1 indicate strong relationships.

The correlation between rainfall and temperature is approximately 0.31, which indicates a weak positive correlation. This means that as temperature increases, rainfall tends to increase slightly, but the relationship is not strong or consistent

```python
# Q 20) What is the correlation between Nitrogen, Phosphorus, Potassium, pH, Rainfall, and Temperature? (Correlation heatmap)
corr = df[['Nitrogen','Phosphorus','Potassium','pH','Rainfall', 'Temperature']].corr()
corr
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(corr,annot = True, cmap ='coolwarm')
plt.show()
```

I have created the correlation matrix of Nitrogen, Phosphorus, Potassium, pH, Rainfall,Temperature. and then visualize it using Heatmap.Heatmap shows which variable is positevly or negativly related.
High positive values (close to 1) mean strong relationship. negative values( close to -1) shows reverse relationship and value near to 0 mean no correlation .
